```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Population of India.csv')
```

```python
data
```

```python
data = data[data['State/UT'] != 'Total (India)']
```

```python
plt.figure(figsize=(12, 8))
bars = plt.barh(data['State/UT'], data['Population[50]'], color='green')

# Add values at the end of each bar
for bar in bars:
    yval = bar.get_y() + bar.get_height() / 2
    xval = bar.get_width()
    plt.text(xval, yval, f'{xval:,.0f}', ha='left', va='center')

plt.xlabel('Population')
plt.title('State-wise Population in India')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
# Calculate the total population for each state
data['Total Population'] = data['Male'] + data['Female']

# Create a bar chart for boys' population percentage for each state
plt.figure(figsize=(15, 10))
bar_width = 0.4
bar_positions = range(len(data))

boys_percentage = (data['Male'] / data['Total Population']) * 100

boys_bars = plt.bar(bar_positions, boys_percentage, width=bar_width, label='Boys', color='blue')

# Add boys' population percentage values above each bar
for boys_bar in boys_bars:
    boys_percentage_value = boys_bar.get_height()
    plt.text(boys_bar.get_x() + boys_bar.get_width() / 2, boys_percentage_value + 1, f'{boys_percentage_value:.1f}%',
             ha='center', va='bottom', color='blue', fontsize=9)

# Add data labels and ticks
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Boys Population Percentage', fontsize=12)
plt.title('Boys Population Percentage by State/UT', fontsize=14)
plt.xticks(bar_positions, data['State/UT'], rotation=45, ha='right', fontsize=10)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Create a bar chart for girls' population percentage for each state
plt.figure(figsize=(15, 10))
bar_width = 0.4
bar_positions = range(len(data))

girls_percentage = (data['Female'] / data['Total Population']) * 100

girls_bars = plt.bar(bar_positions, girls_percentage, width=bar_width, label='Girls', color='pink')

# Add girls' population percentage values above each bar
for girls_bar in girls_bars:
    girls_percentage_value = girls_bar.get_height()
    plt.text(girls_bar.get_x() + girls_bar.get_width() / 2, girls_percentage_value + 1, f'{girls_percentage_value:.1f}%',
             ha='center', va='bottom', color='red', fontsize=9)

# Add data labels and ticks
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Girls Population Percentage', fontsize=12)
plt.title('Girls Population Percentage by State/UT', fontsize=14)
plt.xticks(bar_positions, data['State/UT'], rotation=45, ha='right', fontsize=10)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate the absolute difference between male and female populations
data['Male-Female Difference'] = abs(data['Male'] - data['Female'])

# Sort the DataFrame based on the difference column in descending order
sorted_data = data.sort_values(by='Male-Female Difference', ascending=False).head(10)

# Create a bar chart for the top 10 states with the highest difference
plt.figure(figsize=(12, 8))
bars = plt.barh(sorted_data['State/UT'], sorted_data['Male-Female Difference'], color='red')

# Add values at the end of each bar
for bar in bars:
    yval = bar.get_y() + bar.get_height() / 2
    xval = bar.get_width()
    plt.text(xval, yval, f'{xval:,.0f}', ha='left', va='center')

plt.xlabel('Male-Female Difference in Population')
plt.title('Top 10 States/UTs with Highest Difference in Male and Female Population')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
# Calculate the total population for each state
data['Total Population'] = data['Urban[51]'] + data['Rural[51]']

# Create a bar chart for urban and rural populations side by side for each state
plt.figure(figsize=(15, 10))
bar_width = 0.4
bar_positions = range(len(data))

urban_percentage = (data['Urban[51]'] / data['Total Population']) * 100
rural_percentage = (data['Rural[51]'] / data['Total Population']) * 100

urban_bars = plt.bar(bar_positions, urban_percentage, width=bar_width, label='Urban', color='purple')
rural_bars = plt.bar([pos + bar_width for pos in bar_positions], rural_percentage, width=bar_width, label='Rural', color='lightgreen')

# Add percentage values above each bar
for urban_bar, rural_bar in zip(urban_bars, rural_bars):
    urban_percentage_value = urban_bar.get_height()
    rural_percentage_value = rural_bar.get_height()
    
    plt.text(urban_bar.get_x() + urban_bar.get_width() / 2, urban_percentage_value + 1, f'{urban_percentage_value:.1f}%',
             ha='center', va='bottom', color='blue', fontsize=9)
    
    plt.text(rural_bar.get_x() + rural_bar.get_width() / 2, rural_percentage_value + 1, f'{rural_percentage_value:.1f}%',
             ha='center', va='bottom', color='orange', fontsize=9)

# Add data labels and ticks
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Population Percentage', fontsize=12)
plt.title('Urban vs Rural Population Percentage by State/UT', fontsize=14)
plt.xticks([pos + bar_width / 2 for pos in bar_positions], data['State/UT'], rotation=45, ha='right', fontsize=10)
plt.legend()

plt.tight_layout()
plt.show()
```
